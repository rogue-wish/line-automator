```python
import numpy as np
import matplotlib.pyplot as plt
data=np.loadtxt("he0940-1050_both.dat",unpack=True) #importing the dat file
```

```python
w_data=data[0,:] #wavelength data
f_data=data[1,:] #flux data
e_data=data[2,:] #error data
c_data=data[4,:] #continuum data
nf_data=f_data/c_data #normalized flux data
nf_data_1=f_data/c_data #normalized flux data separate array for doublet overlapping
ne_data=e_data/c_data #normalized error data
```

```python
w_data_1=w_data*(1548.2040/1550.7810) #wavelength multiplied by a factor of ratio of CIV 1548 and CIV1550 line wavelength for overlapping
from scipy.interpolate import interp1d
f3=interp1d(w_data_1,nf_data_1,kind='cubic') #interpolation for bringing nf_data_1 to same wavelength datapoints as original one
```

```python
cw_data=w_data[18097:130325] #selecting data in wavelength range of 3500 to 7000 angstrom
cnf_data=nf_data[18097:130325]
cne_data=ne_data[18097:130325]
cnf_data_1=f3(cw_data)
```

```python
test=5600 #initial value of wavelength range to be tested. At a time, a span of 100 angstrom is tested. Alter values here
i_index=np.where(cw_data<test)[0][-1] #index of initial wavelength
f_index=np.where(cw_data<(test+100))[0][-1]+1 #index of final wavelength
```

```python
ccw_data=cw_data[i_index:f_index] #selecting data in wavelength range as provided by test variable above
ccnf_data=cnf_data[i_index:f_index]
ccne_data=cne_data[i_index:f_index]
ccnf_data_1=cnf_data_1[i_index:f_index]
```

```python
for i in range(len(ccw_data)): 
    if((2*ccnf_data_1[i]-ccnf_data[i]-1)>-5*ccne_data[i] and (2*ccnf_data_1[i]-ccnf_data[i]-1)<5*ccne_data[i]):
        continue #lets the pixels satisfying the doublet condition in error range of 5*sigma be unaltered
    else:
        ccnf_data[i]=1 #changes the value of non satisfying ones to be 1
    
```

```python
plt.figure(figsize=(25,7)) #plot after data has been filtered for doublets
plt.plot(ccw_data,ccnf_data)
plt.xlim(test,test+100)
plt.ylim(0,1.1)
plt.grid()
plt.show()
```

```python
for i in range(len(ccw_data)): 
    if(ccnf_data[i]>(1-5*ccne_data[i])):
        ccnf_data[i]=1 #changes the values of pixels to one which lie between the 5*sigma sandwich
```

```python
plt.figure(figsize=(25,7)) #plot after the data has been cleaned for noise
plt.plot(ccw_data,ccnf_data)
plt.xlim(test,test+100)
plt.ylim(0,1.1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(25,7)) #original overlapping plot without any cleaning
plt.plot(w_data,nf_data)
plt.plot(w_data*(1548.2040/1550.7810),nf_data)
plt.axhline(y = 1, color = 'r', linestyle = '-')
plt.xlim(test,test+100)
plt.ylim(0,1.1)
plt.grid()
plt.show()
```
